```python
#Install + imports + cache
!pip install fastf1 fastf1[plotting]

import os
import fastf1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Enable cache (for now just in Colab; later we can move to Drive)
CACHE_PATH = "/content/fastf1_cache"
os.makedirs(CACHE_PATH, exist_ok=True)
fastf1.Cache.enable_cache(CACHE_PATH)

print("✅ FastF1 ready, cache at:", CACHE_PATH)
```

```python
#Load Silverstone 2023 race (session & basic info)
# Silverstone = British GP
year = 2023
event_name = "British"  #  "British" for Silverstone
session_type = "Race"

session = fastf1.get_session(year, event_name, session_type)
session.load()

print(session.event['EventName'], session.event['EventDate'])
session.results.head()
```

```python
#Build core race tables (laps, stints, tyres, pits)(Laps with driver names & compounds)
laps = session.laps.copy()

# Add driver names and team
laps = laps.merge(session.results[['DriverNumber', 'Abbreviation', 'TeamName']],
                  left_on='DriverNumber', right_on='DriverNumber', how='left')

# Keep key columns
lap_cols = [
    'DriverNumber', 'Abbreviation', 'TeamName',
    'LapNumber', 'LapTime', 'Sector1Time', 'Sector2Time', 'Sector3Time',
    'Stint', 'Compound', 'TyreLife', 'FreshTyre', 'TrackStatus',
    'PitOutTime', 'PitInTime', 'IsPersonalBest'
]
laps_clean = laps[lap_cols].copy()

# Convert LapTime to seconds (float)
laps_clean['LapTimeSeconds'] = laps_clean['LapTime'].dt.total_seconds()

laps_clean.head()
```

```python
#Pit stops table
# Simple pit-stops table: any lap with pit in/out
pits = laps_clean[~laps_clean['PitInTime'].isna() | ~laps_clean['PitOutTime'].isna()].copy()
pits = pits[['DriverNumber', 'Abbreviation', 'TeamName', 'LapNumber', 'PitInTime', 'PitOutTime']]
pits.head()
```

```python
#Filter VER & HAM race laps
target_drivers = ["VER", "HAM"]

race_laps = laps_clean[
    laps_clean['Abbreviation'].isin(target_drivers)
].copy()

# Remove laps without laptime or clearly out/in laps
race_laps = race_laps[~race_laps['LapTime'].isna()]
race_laps = race_laps[race_laps['LapTimeSeconds'] > 50]  # crude filter for out/in laps if needed

race_laps.head()
```

```python
#Race pace line chart (VER vs HAM)

plt.figure(figsize=(10,5))

for drv in target_drivers:
    drv_laps = race_laps[race_laps['Abbreviation'] == drv]
    plt.plot(drv_laps['LapNumber'], drv_laps['LapTimeSeconds'], marker='o', label=drv)

plt.title("Race Pace – Silverstone 2023 (VER vs HAM)")
plt.xlabel("Lap Number")
plt.ylabel("Lap Time (s)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Telemetry comparison (fastest laps VER vs HAM)
```

```python
#Get fastest laps & telemetry
import fastf1.plotting
fastf1.plotting.setup_mpl()

# Pick fastest lap for each driver
lap_ver = session.laps.pick_driver("VER").pick_fastest()
lap_ham = session.laps.pick_driver("HAM").pick_fastest()

tel_ver = lap_ver.get_telemetry().add_distance()
tel_ham = lap_ham.get_telemetry().add_distance()

tel_ver.head(), tel_ham.head()
```

```python
#Speed vs Distance plot
plt.figure(figsize=(12,5))
plt.plot(tel_ver['Distance'], tel_ver['Speed'], label="VER")
plt.plot(tel_ham['Distance'], tel_ham['Speed'], label="HAM")

plt.title("Fastest Lap Telemetry – Speed vs Distance (Silverstone 2023)")
plt.xlabel("Distance (m)")
plt.ylabel("Speed (km/h)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Simple driver metrics for race (VER vs HAM)
# Aggregate stats per driver
summary = (
    race_laps
    .groupby('Abbreviation')
    .agg(
        avg_lap_time=('LapTimeSeconds', 'mean'),
        median_lap_time=('LapTimeSeconds', 'median'),
        std_lap_time=('LapTimeSeconds', 'std'),
        laps_count=('LapNumber', 'count')
    )
    .reset_index()
)

summary
```

```python
#Sector Performance & Lap Delta Analysis (VER vs HAM — Silverstone 2023)
```

```python
#This will show you where Verstappen was faster than Hamilton:

#Sector 1 (high-speed sweepers)

#Sector 2 (technical middle)

#Sector 3 (slow corners + straights)

#…and how much time each driver gained/lost per lap.
```

```python
#Building Sector Table (S1, S2, S3)
```

```python
# Keeping only VER and HAM laps with valid sector times
sector_df = laps_clean[
    (laps_clean['Abbreviation'].isin(['VER', 'HAM'])) &
    (~laps_clean[['Sector1Time', 'Sector2Time', 'Sector3Time']].isna().any(axis=1))
][['LapNumber', 'Abbreviation', 'Sector1Time', 'Sector2Time', 'Sector3Time']].copy()

# Convert to seconds
sector_df['S1'] = sector_df['Sector1Time'].dt.total_seconds()
sector_df['S2'] = sector_df['Sector2Time'].dt.total_seconds()
sector_df['S3'] = sector_df['Sector3Time'].dt.total_seconds()

sector_df.head()
```

```python
#Calculating Per-Lap Delta (VER minus HAM)

# Pivot to make VER and HAM separate columns
pivot_s1 = sector_df.pivot(index='LapNumber', columns='Abbreviation', values='S1')
pivot_s2 = sector_df.pivot(index='LapNumber', columns='Abbreviation', values='S2')
pivot_s3 = sector_df.pivot(index='LapNumber', columns='Abbreviation', values='S3')

# Delta = HAM - VER (positive means VER is faster)
pivot_s1['Delta_S1'] = pivot_s1['HAM'] - pivot_s1['VER']
pivot_s2['Delta_S2'] = pivot_s2['HAM'] - pivot_s2['VER']
pivot_s3['Delta_S3'] = pivot_s3['HAM'] - pivot_s3['VER']

sector_delta = pd.concat([pivot_s1['Delta_S1'], pivot_s2['Delta_S2'], pivot_s3['Delta_S3']], axis=1)
sector_delta.head()
```

```python
#Plotting Sector Deltas

plt.figure(figsize=(12,8))

plt.plot(sector_delta.index, sector_delta['Delta_S1'], label="Sector 1")
plt.plot(sector_delta.index, sector_delta['Delta_S2'], label="Sector 2")
plt.plot(sector_delta.index, sector_delta['Delta_S3'], label="Sector 3")

plt.axhline(0, color='black', linestyle='--', linewidth=1)

plt.title("Sector Performance Delta (HAM - VER)\nPositive = VER Faster")
plt.xlabel("Lap Number")
plt.ylabel("Sector Delta (seconds)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Above 0 → VER was faster

#Below 0 → HAM was faster
```

```python
#Total Lap Delta (overall pace per lap)
lap_delta = laps_clean[
    laps_clean['Abbreviation'].isin(['VER', 'HAM'])
][['LapNumber', 'Abbreviation', 'LapTimeSeconds']]

lap_delta = lap_delta.pivot(index='LapNumber', columns='Abbreviation', values='LapTimeSeconds')
lap_delta['Delta_Lap'] = lap_delta['HAM'] - lap_delta['VER']

lap_delta.head()
```

```python
#Plot
plt.figure(figsize=(10,5))
plt.plot(lap_delta.index, lap_delta['Delta_Lap'], label='Total Lap Delta', marker='o')

plt.axhline(0, color='black', linestyle='--')
plt.title("Total Lap Time Delta (HAM - VER)\nPositive = VER Faster")
plt.xlabel("Lap Number")
plt.ylabel("Lap Delta (seconds)")
plt.tight_layout()
plt.show()
```

```python
#per-lap cumulative race time for each driver
import pandas as pd
import numpy as np

# Working on a copy
race_laps = laps_clean.copy()

# Filtering only proper race laps (no NaN laptime)
race_laps = race_laps.dropna(subset=['LapTimeSeconds'])

# Computing cumulative race time per driver
race_laps['CumTime'] = (
    race_laps
    .sort_values(['Abbreviation', 'LapNumber'])
    .groupby('Abbreviation')['LapTimeSeconds']
    .cumsum()
)

race_laps[['Abbreviation', 'LapNumber', 'Position', 'CumTime']].head()
```

```python
# Restrict to race laps only (Position not null)
grid = race_laps.dropna(subset=['Position']).copy()
grid['Position'] = grid['Position'].astype(int)

grid_simple = grid[['LapNumber', 'Abbreviation', 'Position', 'CumTime']].copy()
grid_simple.head()
```

```python
# Car ahead: same lap, position - 1
ahead = grid_simple.copy()
ahead['Position'] = ahead['Position'] + 1  # so when we merge, we match "our" row at Position, and ahead is Position-1
ahead = ahead.rename(columns={
    'Abbreviation': 'AheadDriver',
    'CumTime': 'AheadCumTime'
})

# Car behind: same lap, position + 1
behind = grid_simple.copy()
behind['Position'] = behind['Position'] - 1
behind = behind.rename(columns={
    'Abbreviation': 'BehindDriver',
    'CumTime': 'BehindCumTime'
})

# Merge both onto main grid_simple
racecraft = (
    grid_simple
    .merge(ahead[['LapNumber', 'Position', 'AheadDriver', 'AheadCumTime']],
           on=['LapNumber', 'Position'], how='left')
    .merge(behind[['LapNumber', 'Position', 'BehindDriver', 'BehindCumTime']],
           on=['LapNumber', 'Position'], how='left')
)

racecraft.head()
```

```python
# Time gaps in seconds
racecraft['GapAhead'] = racecraft['AheadCumTime'] - racecraft['CumTime']
racecraft['GapBehind'] = racecraft['CumTime'] - racecraft['BehindCumTime']

# Keep only VER & HAM for now
racecraft_vh = racecraft[racecraft['Abbreviation'].isin(['VER', 'HAM'])].copy()

racecraft_vh[['Abbreviation', 'LapNumber', 'Position', 'GapAhead', 'GapBehind']].head()
```

```python
ATTACK_THRESHOLD = 1.0  # seconds
DEFENSE_THRESHOLD = 1.0

def classify_zone(row):
    attack = (row['GapAhead'] is not None) and (not pd.isna(row['GapAhead'])) and (row['GapAhead'] <= ATTACK_THRESHOLD)
    defense = (row['GapBehind'] is not None) and (not pd.isna(row['GapBehind'])) and (row['GapBehind'] <= DEFENSE_THRESHOLD)

    if attack and defense:
        return 'Attack & Defense'
    elif attack:
        return 'Attack'
    elif defense:
        return 'Defense'
    else:
        return 'Neutral'

racecraft_vh['Zone'] = racecraft_vh.apply(classify_zone, axis=1)

racecraft_vh[['Abbreviation', 'LapNumber', 'Position', 'GapAhead', 'GapBehind', 'Zone']].head()
```

```python
import matplotlib.pyplot as plt

zone_colors = {
    'Attack': 'tab:red',
    'Defense': 'tab:blue',
    'Attack & Defense': 'tab:purple',
    'Neutral': 'tab:gray'
}

plt.figure(figsize=(12,4))

for drv in ['VER', 'HAM']:
    drv_data = racecraft_vh[racecraft_vh['Abbreviation'] == drv]
    y = 1 if drv == 'VER' else 0  # two horizontal lines

    for _, row in drv_data.iterrows():
        plt.scatter(row['LapNumber'], y,
                    color=zone_colors[row['Zone']],
                    s=50)

plt.yticks([0,1], ['HAM', 'VER'])
plt.xlabel("Lap Number")
plt.title("Racecraft Zones – Attack / Defense / Neutral (VER vs HAM, Silverstone 2023)")
plt.grid(axis='x', linestyle='--', alpha=0.3)

# Legend
handles = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                      markerfacecolor=color, markersize=8)
           for label, color in zone_colors.items()]
plt.legend(handles=handles, title="Zone", bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
